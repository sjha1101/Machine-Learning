```python
#common types of blur
# 1.Box blur
# 2.Gaussian blur
# 3.median blur
# 4.motion blur
# (A)Horizontal (B) Vertial
```

```python
#1.Box blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('nature.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
box=cv2.blur(img,(10,10))#We can have any positive number
plt.subplot(1,2,1),plt.imshow(img),plt.title('Source image')
plt.subplot(1,2,2),plt.imshow(box),plt.title('Box blur image')
```

```python
#2.Gaussian blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('nature.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gaus=cv2.GaussianBlur(img,(11,11),0) #We can have any positive odd number & 0 means a natural blur effect on the image
plt.subplot(1,2,1),plt.imshow(img),plt.title('Source image')
plt.subplot(1,2,2),plt.imshow(gaus),plt.title('Gaussian blur image')
```

```python
#3.Median blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('nature.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
med=cv2.medianBlur(img,11)
plt.subplot(1,2,1),plt.imshow(img),plt.title('Source image')
plt.subplot(1,2,2),plt.imshow(med),plt.title('Median blur image')
```

```python
# 4.Motion blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('nature.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
size=30
# (A)Horizontal
k1=np.zeros((size,size))
k1[int(size/2),:]=np.ones(size)
k1=k1/size

#(b)VERTICAL
k2=np.zeros((size,size))
k2[:,int(size/2)]=np.ones(size)
k2=k2/size

#applying the kernal
horizontal=cv2.filter2D(img,-1,k1)
vertical=cv2.filter2D(img,-1,k2)

plt.subplot(1,3,1),plt.imshow(img),plt.title('Source image')
plt.subplot(1,3,2),plt.imshow(horizontal),plt.title('Horizontal blur image')
plt.subplot(1,3,3),plt.imshow(vertical),plt.title('vertical blur image')
```
