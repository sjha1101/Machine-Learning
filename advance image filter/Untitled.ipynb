```python
#Applying advance image filters
#1). Bilateral(for the advance smoothing effect.)
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('face.png')
bilateral=cv2.bilateralFilter(img,d=9,sigmaColor=35,sigmaSpace=85)

plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('Source image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(bilateral,cv2.COLOR_BGR2RGB)),plt.title('with somothing effect'),plt.axis('off')
```

```python
# 2)Unsharp 
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('mountain.jpg')

#blur the image
gauss=cv2.GaussianBlur(img,(9,9),10)

#unsharp
unsharp=cv2.addWeighted(img,1.5,gauss,-0.5,0)

plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(unsharp,cv2.COLOR_BGR2RGB)),plt.title('unsharp image'),plt.axis('off')
```

```python
# 3)scharr(for edge detection)
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('zibra.jpg',0)

#Apply scvharr filter
scharrx=cv2.Scharr(img,cv2.CV_64F,1,0) #vertical scan
scharry=cv2.Scharr(img,cv2.CV_64F,0,1) #horizontal scan

#convert it to unit8 for display
scharrx=cv2.convertScaleAbs(scharrx)
scharry=cv2.convertScaleAbs(scharry)

#combine both directions
scharr_combined=cv2.addWeighted(scharrx,0.10,scharry,0.2,0)

plt.subplot(2,2,1),plt.imshow(img,cmap='gray'),plt.title('Source image'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(scharrx,cmap='gray'),plt.title('scharrx'),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(scharry,cmap='gray'),plt.title('scharry'),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(scharr_combined,cmap='gray'),plt.title('scharr_combined'),plt.axis('off')
```
