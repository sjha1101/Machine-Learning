```python
#Erosian & dilation
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Loading image
img = cv2.imread('0to9.jpg', 0)
img = cv2.bitwise_not(img)

# Defining a kernel
kernel = np.ones((3, 3), np.uint8)

# Applying erosion
erosion = cv2.erode(img, kernel, iterations=1)

# Applying dilation
dilation = cv2.dilate(img, kernel, iterations=1)

# Displaying the images
plt.subplot(1, 3, 1), plt.imshow(img, cmap='gray'), plt.title('Source'), plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(erosion, cmap='gray'), plt.title('Erode effect'), plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(dilation, cmap='gray'), plt.title('Dilate effect'), plt.axis('off')
```
