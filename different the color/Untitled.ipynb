```python
# Showing effect of Y (Luminous), Cr (Chrominance red) and Cb (Chrominance blue)
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('rose.jpg')

# Convert it to YCrCb
ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

# Split channels
y, cr, cb = cv2.split(ycrcb)

# Plotting images
plt.figure(figsize=(6,6))
plt.subplot(2,2,1), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Source'), plt.axis('off')
plt.subplot(2,2,2), plt.imshow(y, cmap='gray'), plt.title('Y (Luminance)'), plt.axis('off')
plt.subplot(2,2,3), plt.imshow(cr, cmap='gray'), plt.title('Cr (Chrominance red)'), plt.axis('off')
plt.subplot(2,2,4), plt.imshow(cb, cmap='gray'), plt.title('Cb (Chrominance blue)'), plt.axis('off')
```
