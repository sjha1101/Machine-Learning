```python
# Edge detection
# A)sobel
# B)laplacian
# C) canny
# A)sobel (img,depth,x,y,kernal_size)
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('cat.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#Edges work on intensity of the color not on the color

sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,25)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,25)
sobel_combined=cv2.magnitude(sobelx,sobely)

plt.subplot(2,2,1),plt.imshow(img),plt.title('source img'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(sobelx,cmap='gray'),plt.title('Sobel X'),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(sobely,cmap='gray'),plt.title('Sobel y'),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(sobel_combined,cmap='gray'),plt.title('Sobel image'),plt.axis('off')
```

```python
# B)laplacian
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('cat.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#Edges work on intensity of the color not on the color

#Gaussian blur
Gaussian=cv2.GaussianBlur(img,(3,3),0)

#Laplacian(src,depth,k_size)
lalplacian=cv2.Laplacian(img,cv2.CV_64F,ksize=3)
lalplacian_after_blur=cv2.Laplacian(Gaussian,cv2.CV_64F,ksize=3)

plt.figure(figsize=(10,6))

plt.subplot(1,3,1),plt.imshow(img),plt.title('source image'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(lalplacian,cmap='gray'),plt.title('lalplacian image'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(lalplacian_after_blur,cmap='gray'),plt.title('lalplacian_after_blur image'),plt.axis('off')
```

```python
# C)canny(src,threshold1,threshold2)
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('cat.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#Edges work on intensity of the color not on the color

#Gaussian blur
Gaussian=cv2.GaussianBlur(img,(3,3),0)

canny=cv2.Canny(Gaussian,threshold1=50,threshold2=150)

plt.figure(figsize=(10,6))

plt.subplot(1,3,1),plt.imshow(img),plt.title('source image'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(canny,cmap='gray'),plt.title('canny image'),plt.axis('off')
```
